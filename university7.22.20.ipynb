```python
# mean
utype = university.groupby('Type')
utype2 = utype.mean()
utype2
```

```python
# median
utype3 = utype.median()
utype3
```

```python
# order by average size 
utype2.sort_values(by=['Size'])
```

```python
# putting together mean and median by type and sorts values mean asc

utype2.join(utype3, how='inner', lsuffix= 'mean', rsuffix = 'median').sort_values(by=['Sizemean'])
```

```python
# split address 
addresses = university['address']
type(addresses), addresses
```

```python
# create new columns IN the university dataframe and assigning them 
# rsplit function splits from right hand instead of the left 
# this splits twice on the first two commas

university[['street', 'city', "state"]] = university.address.str.rsplit(',',2, expand=True)
university
```

```python
# split state column on a space 
university[["state2", "zip"]] = university.state.str.rsplit(' ',1, expand=True)
university
```

```python
adresses2 = addresses[2].str.split(" ", expand=True)
adresses2
```

```python
# rename column in addresses2
adresses2 = adresses2.rename(columns={0:'nnon',1:'state', 2:'zip', 3:'non1', 4:'non2', 5:'non3'})
adresses2
```

```python
university
```

```python
#  drop 
uni = university.drop(['address', 'Website', 'Type', 'street', 'city', 'state',
       'zip'], axis=1)
uni
```

```python
# concat uni + addresses for mean + median by state 
# reference: data1 = pd.concat([data, names], axis=1)

bystate = pd.concat([uni, adresses2], axis =1)
bystate
```

```python
# drop zip-5
bystate2 = bystate.drop(['nnon','zip', 'non1', 'non2', 'non3'],axis=1)
bystate2
```

```python
# mean by state 
state2 = bystate.groupby('state').mean()
state2
```

```python
state2.head(60)
```

```python
# median by state 
state3 = bystate.groupby('state').median()
state3
```

```python
# mean by name 
name2 = bystate.groupby('name').mean()
name2
```

```python
# median by name 
name3 = bystate.groupby('name').median()
name3
```

```python
university['Public'] = np.where(university['Type'].str.contains('Private')>0, 'Private', 'Public')
university
```

```python
university['For_profit'] = np.where(university.Type.str.contains(' for')>0, 'For-Profit', 'NFP')
university
```

```python
university['Primarily_Associates?'] = np.where(university.Type.str.contains('prim')>0 , 1, 0)
university
```

```python
university.groupby(['Public', 'Primarily_Associates?']).aggregate([np.mean, np.median])
```
